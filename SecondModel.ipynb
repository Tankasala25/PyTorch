```python
import torch
import torch.nn as nn

x=torch.tensor([[1],[2],[3],[4]],dtype=torch.float32)
y=torch.tensor([[3],[6],[9],[12]],dtype=torch.float32)

class LinearModel(nn.Module):
  def __init__(self):
    super().__init__()
    self.fc1=nn.Sequential(
        nn.Linear(in_features=1,out_features=1,bias=False)
    )

  def forward(self,x):
    return self.fc1(x)

model_1=LinearModel()
loss =nn.MSELoss()
optimizer=torch.optim.SGD(model_1.parameters(),lr=0.01)

epochs=100
for epoch in range(epochs):

  #forward function
  y_pred=model_1(x)

  #calculate the loss
  l=loss(y_pred,y)

  #calculate the gradient
  l.backward()

  #optimize the gradient with parameters
  optimizer.step()

  #get the gradients to zero
  optimizer.zero_grad()

  if epoch%10==0:
    print(f"{epoch} : loss :{l.item():.8f}, weights : {model_1.fc1[0].weight.data}")

```

```python
from pathlib import Path

#Creating a model Directory path

MODEL_Path=Path("Models")
MODEL_Path.mkdir(parents=True,exist_ok=True)

#Creating a model save path

Model_name= "Pytorch_SecondModel.pth"
Model_save_Path=MODEL_Path/Model_name

#Saving a Model
print(f"Saving model to path :{Model_save_Path}")
torch.save(obj=model_1.state_dict(),f=Model_save_Path)
```

```python
model_loaded_state1=LinearModel()

model_loaded_state1.load_state_dict(torch.load(f=Model_save_Path))
```
