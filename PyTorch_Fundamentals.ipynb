### Batch multiplication

Only the last two dims are used for matrix multiplication (m×n × n×p → m×p).

All earlier dims are batch dims.

Batch dims must either match or one of them must be 1 (for broadcasting).

```python
A = torch.randn(1, 3, 4)  # (batch=1)
B = torch.randn(3, 4, 5)  # (batch=2)
C = torch.matmul(A, B)
print(C.shape)
```

```python
A = torch.randn(2, 3, 4)  # (batch=1)
B = torch.randn(3, 4, 5)  # (batch=2)
C = torch.matmul(A, B)
print(C.shape)
```

# Tensor Operation Differences (Short Notes)

### Differences Between `reshape`, `transpose`, and `permute`

| Function | Purpose | What It Does | Example Change | Notes |
|-----------|----------|---------------|----------------|-------|
| **`reshape()`** | Change shape | Rearranges elements into a new shape (no axis change) | `(2, 3, 4)` → `(3, 8)` | Same data order, just reshaped |
| **`transpose(dim0, dim1)`** | Swap two axes | Exchanges two dimensions only | `(2, 3, 4)` → `(3, 2, 4)` | Works for 2D/3D tensors |
| **`permute(dims)`** | Reorder all axes | Changes the order of *all* dimensions | `(2, 3, 4)` → `(4, 2, 3)` | Generalized version of transpose |

---

💡 **In short:**
- `reshape` → changes **shape**  
- `transpose` → swaps **two axes**  
- `permute` → **reorders all axes**

###Reproducibility

Reproducibility means getting the same results every time you run your code —
even if your model uses random initialization, random shuffling, or random noise.

In PyTorch, you control this mainly by setting a random seed.

```python
# Same seed → same results
torch.manual_seed(42)
r1 = torch.rand(2, 3)

torch.manual_seed(42)
r2 = torch.rand(2, 3)



# Different seed → different results
torch.manual_seed(7)
r3 = torch.rand(2, 3)

# Any integer works (1729 is just an example)
torch.manual_seed(1729)
r4 = torch.rand(2, 3)

r5=torch.rand(2,3)

torch.manual_seed(1729)

r6= torch.rand(2, 3)

r7=torch.rand(2,3)

print("Seed 42 - first:\n", r1)
print("\nSeed 42 - second (identical):\n", r2)
print(r1==r2)
print("\nSeed 7:\n", r3)
print("\nSeed 1729:\n", r4)
print("\nSeed 1729:\n", r5)
print("\nSeed 1729:\n", r6)
print("\nSeed 1729:\n", r7)
print(r4==r6)
print(r5==r7)
```
